```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import os
import shutil
from sklearn.metrics import accuracy_score

%matplotlib inline
seed = 0
np.random.seed(seed)

tf.random.set_seed(seed)

os.environ['PATH'] = os.environ['XILINX_VITIS'] + '/bin:' + os.environ['PATH']

model_name = "MNIST_MLP_HGQ_DynamicTraining"
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
X_test = X_test.astype("float32") / 255
y_test = keras.utils.to_categorical(y_test, 10)
```

```python
from keras.models import load_model
import hgq.layers


model = load_model(f"MNIST_MLP_HGQ_DynamicTraining_model.keras")
score = model.evaluate(X_test, y_test)
```

```python
model.summary()
```

```python
import hls4ml

output_dir = f"hls4ml_prj_{model_name}"
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
    
hls_config = hls4ml.utils.config_from_keras_model(model, granularity='name')

#hls_config['Model']['Strategy'] = 'Distributed Arithmetic'
#proj_name = f"{str(model_to_test)}_{str(model_revision)}_hls4ml_prj_{str(hls4ml_revision)}"

hls_model = hls4ml.converters.convert_from_keras_model(
    model,    
    backend='vitis',
    hls_config=hls_config,
    #io_type='io_stream',
    #proj_name = proj_name,
    output_dir=output_dir, 
    board       = 'kv260',
    part='xck26-sfvc784-2LV-c',
    clock_period='5',
)
hls_model.compile()
```

```python
hls_model.build(csim=False, synth=True, cosim=False)
```
