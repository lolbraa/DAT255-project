```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
datapath = f"MNIST_HGQ_DynamicTraining_log.csv"

df = pd.read_csv(datapath, sep=';',index_col=0) #https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html

#df.head
df[:10]
```

```python
df.index
```

```python
plt.figure()
plt.plot(df.index, 'accuracy', data=df)
#plt.plot(df.index, 'ebops', data=df, scaley=False)
#plt.plot(df.index, 'learning_rate', data=df)
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Epochs')
color = 'tab:orange'
ax1.plot(df.index, 'accuracy', data=df, color=color)
ax1.set_ylabel('Accuracy', color=color)

color = 'tab:blue'
ax2 = ax1.twinx()
ax2.set_ylabel('Ebops', color=color)
ax2.semilogy(df.index, 'ebops', data=df, color=color) # https://matplotlib.org/stable/gallery/scales/log_demo.html
#ax2.plot(df.index, 'ebops', data=df, color=color)

# Annotations https://matplotlib.org/stable/gallery/text_labels_and_annotations/annotate_transform.html

# Minimalistic https://practicaldatascience.org/notebooks/class_5/week_1/2.2.2_making_plots_pretty_2.html

plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Ebops')
color = 'tab:orange'
ax1.plot('ebops', 'accuracy', data=df, color=color)
ax1.set_ylabel('Accuracy', color=color)

# Annotations https://matplotlib.org/stable/gallery/text_labels_and_annotations/annotate_transform.html

# Minimalistic https://practicaldatascience.org/notebooks/class_5/week_1/2.2.2_making_plots_pretty_2.html

plt.show()
```
